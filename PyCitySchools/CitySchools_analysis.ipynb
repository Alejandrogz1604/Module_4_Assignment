```python
import pandas as pd
from pathlib import Path
import numpy as np

```

```python
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")
```

```python
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
school_data_complete = pd.merge(student_data, school_data, how = "left", on = ["school_name", "school_name"])
```

```python
individual_school_list = school_data_complete['school_name'].unique()
individual_school_count = len(individual_school_list)

individual_school_count
```

```python
student_id_list = school_data_complete['Student ID'].unique()
individual_student_count = len(student_id_list)

individual_student_count
```

```python
total_budget = school_data_complete.drop_duplicates(subset = 'budget', keep = 'first')['budget'].sum()

total_budget 
```

```python
average_math_score = school_data_complete['math_score'].mean()
average_math_score
```

```python
average_reading_score = school_data_complete['reading_score'].mean()
average_reading_score
```

```python
passing_math_count = school_data_complete[(school_data_complete["math_score"]>=70)].count()["student_name"]
math_passing_percentage = passing_math_count / individual_student_count * 100
math_passing_percentage
```

```python
passing_reading_count = school_data_complete[(school_data_complete["reading_score"]>=70)].count()["student_name"]
reading_passing_percentage = passing_reading_count / individual_student_count * 100
reading_passing_percentage
```

```python
passing_overall_count = school_data_complete[(school_data_complete["reading_score"]>=70) & (school_data_complete["math_score"]>=70)].count()["student_name"]
overall_passing_percentage = passing_overall_count / individual_student_count * 100
overall_passing_percentage
```

```python
district_summary = pd.DataFrame({"Total Schools": [individual_school_count], 
                                 "Total Students": [individual_student_count],
                                 "Total Budget": [total_budget],
                                 "Average Math Score": [average_math_score],
                                 "Average Reading Score": [average_reading_score],
                                 "% Passing Math": [math_passing_percentage],
                                 "% Passing Reading": [reading_passing_percentage],
                                 "% Overall Passing": [overall_passing_percentage]})




district_summary
```

```python
individual_school_list = school_data_complete.drop_duplicates(subset = "school_name", keep = "first").sort_values(by = "school_name", ascending = True)
school_type = individual_school_list["type"]
```

```python
students_per_school = school_data_complete.groupby("school_name")["Student ID"].size()
budget_per_school = school_data_complete.groupby("school_name")["budget"].mean()
```

```python
budget_per_capita = budget_per_school / students_per_school
```

```python
per_school_math = school_data_complete.groupby("school_name")["math_score"].mean()
per_school_reading = school_data_complete.groupby("school_name")["reading_score"].mean()
```

```python
students_passing_math = school_data_complete[(school_data_complete["math_score"] >= 70)]
students_passing_math_byschool = students_passing_math.groupby("school_name").size()
```

```python
students_passing_reading = school_data_complete[(school_data_complete["reading_score"] >= 70)]
students_passing_reading_byschool = students_passing_reading.groupby("school_name").size()
```

```python
students_passing_overall = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]
students_passing_overall_byschool = students_passing_overall.groupby("school_name").size()
```

```python
percentage_passing_math_byschool = students_passing_math_byschool / students_per_school * 100
percentage_passing_reading_byschool = students_passing_math_byschool / students_per_school * 100
percentage_passing_overall_byschool = students_passing_overall_byschool / students_per_school * 100
```

```python
per_school_summary = pd.DataFrame( 
                                  {"School Type": school_type,
                                   "Total Students": students_per_school.to_list(),
                                   "Total School Budget": budget_per_school.to_list(), 
                                   "Per Student Budget": budget_per_capita.to_list(), 
                                   "Average Math Score": per_school_math.to_list(), 
                                   "Average Reading Score": per_school_reading.to_list(), 
                                   "% Passing Math": percentage_passing_math_byschool.to_list(), 
                                   "% Passing Reading": percentage_passing_reading_byschool.to_list(), 
                                   "% Passing Overall": percentage_passing_overall_byschool.to_list()
                                  }).set_index(keys = students_per_school.index)


per_school_summary
```

```python
top_schools = per_school_summary.sort_values(by = "% Passing Overall", ascending = False)
top_schools.head()
```

```python
top_schools = per_school_summary.sort_values(by = "% Passing Overall", ascending = True)
top_schools.head()
```

```python
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

ninth_graders_math_scores = ninth_graders.groupby("school_name")["math_score"].mean()
tenth_graders_math_scores = tenth_graders.groupby("school_name")["math_score"].mean()
eleventh_graders_math_scores = eleventh_graders.groupby("school_name")["math_score"].mean()
twelfth_graders_math_scores = twelfth_graders.groupby("school_name")["math_score"].mean()

math_scores_by_grade = pd.DataFrame({"9th": ninth_graders_math_scores.to_list()  ,
                                     "10th": tenth_graders_math_scores.to_list() , 
                                     "11th": eleventh_graders_math_scores.to_list() , 
                                     "12th": twelfth_graders_math_scores.to_list(), }).set_index(ninth_graders_math_scores.index)

math_scores_by_grade
                                     
```

```python
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

ninth_graders_reading_scores = ninth_graders.groupby("school_name")["reading_score"].mean()
tenth_graders_reading_scores = tenth_graders.groupby("school_name")["reading_score"].mean()
eleventh_graders_reading_scores = eleventh_graders.groupby("school_name")["reading_score"].mean()
twelfth_graders_reading_scores = twelfth_graders.groupby("school_name")["reading_score"].mean()

reading_scores_by_grade = pd.DataFrame({"9th": ninth_graders_reading_scores.to_list()  ,
                                     "10th": tenth_graders_reading_scores.to_list() , 
                                     "11th": eleventh_graders_reading_scores.to_list() , 
                                     "12th": twelfth_graders_reading_scores.to_list(), }).set_index(ninth_graders_reading_scores.index)

reading_scores_by_grade
                                     
```

```python
spending_bins = [0, 585, 630, 645, 680]
labels =  ["<$585", "$585-630", "$630-645", "$645-680"]

school_spending_df = per_school_summary.copy()

school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"], 
                                                             spending_bins, labels = labels,
                                                             include_lowest = True)

school_spending_df
```

```python
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
spending_passing_overall = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Overall"].mean()
```

```python
spending_summary = pd.DataFrame({"Average Math Score": spending_math_scores.to_list(),
                                "Average Reading Score": spending_reading_scores.to_list(),
                                "% Passing Math": spending_passing_math.to_list() ,
                                "% Passing Reading": spending_passing_reading.to_list() ,
                                "% Overall Passing": spending_passing_overall.to_list() }).set_index(spending_math_scores.index)

spending_summary
```

```python
size_bins = [0, 1000, 2000, 5000]
labels =  ["Small(<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

per_school_summary["School Size"] = pd.cut(per_school_summary["Total Students"], 
                                                             size_bins, labels = labels,
                                                             include_lowest = True)

per_school_summary

```

```python
size_math_scores = per_school_summary.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["% Passing Math"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["% Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["% Passing Overall"].mean()
```

```python
size_summary = pd.DataFrame({"Average Math Score": size_math_scores.to_list(),
                                "Average Reading Score": size_reading_scores.to_list(),
                                "% Passing Math": size_passing_math.to_list() ,
                                "% Passing Reading": size_passing_reading.to_list() ,
                                "% Overall Passing": size_overall_passing.to_list() }).set_index(size_math_scores.index)

size_summary
```

```python
average_math_score_by_type = per_school_summary.groupby(["School Type"])["Average Math Score"].mean()
average_reading_score_by_type = per_school_summary.groupby(["School Type"])["Average Reading Score"].mean()
average_percent_passing_math_by_type = per_school_summary.groupby(["School Type"])["% Passing Math"].mean()
average_percent_passing_reading_by_type = per_school_summary.groupby(["School Type"])["% Passing Reading"].mean()
average_percent_passing_overall_by_type = per_school_summary.groupby(["School Type"])["% Passing Overall"].mean()

```

```python
type_summary = pd.DataFrame({"Average Math Score": average_math_score_by_type.to_list(),
                                "Average Reading Score": average_reading_score_by_type.to_list(),
                                "% Passing Math": average_percent_passing_math_by_type.to_list() ,
                                "% Passing Reading": average_percent_passing_reading_by_type.to_list() ,
                                "% Overall Passing": average_percent_passing_overall_by_type.to_list() }).set_index(average_math_score_by_type.index)

type_summary
```
